**Publication to cite**: Romitti, Y., I. Sue Wing, K.R. Spangler, and G.A. Wellenius. Inequality in the availability of residential air conditioning across 115 US metropolitan areas. Working paper. 2022.

## Why does this matter?
Air-conditioning is widely understood as an important protective factor against heat-related health problems. During heat-waves, access to air-conditioning can make all the difference to a person's risk of mortality/morbidity. However, access the air-conditioning at home is not equitably distributed throughout neighborhoods. When considering the vulnerability of Chicago communities to extreme weather, the prevalence of AC access in the area is an important metric.

We should also remember that the mere presence of an air-conditioning unit does not guarantee true access. Not all AC-units are operational, and households experiencing income-deprivation may not have the monetary resources to pay for extra electricity costs associated with AC usage. Thus, while the presence of air-conditioning (what this paper calls "access") is important, it is not the only factor affecting a household's adaptive capacity.

## How to find a specific region when working with census data
Chicago is part of the Chicago-Naperville-Elgin, IL-IN-WI core-based statistical area (CBSA 16980). Within this CBSA are several counties. We can identify these counties by their Federal Information Processing System (FIPS) Codes.

As explained in the source below: "FIPS codes are numbers which uniquely identify geographic areas... State-level FIPS codes have two digits, county-level FIPS codes have five digits of which the first two are the FIPS code of the state to which the county belongs."

**Source**: https://transition.fcc.gov/oet/info/maps/census/fips/fips.txt#:~:text=Federal%20Information%20Processing%20System%20

Note that `ac_df` has already been filtered to CBSA 16980. Here are the FIPS codes for all the counties within this CBSA.

To plot the AC-probability values for all of these census tracts, we need to get shapefile information for each one. Those shapefiles can be downloaded for individual counties from the US Census Bureau's TIGER/Line Shapefiles database. I have already done the work for you, but here are the steps I followed. The resulting data lives in `crocus_heat/datasets/boundaries_geospatial_clean/tl_2010_tracts/`.
- Go to https://www.census.gov/geographies/mapping-files/time-series/geo/tiger-line-file.html
- Pick your desired year. I chose 2010, because this is what Romitti et. Al. (2022) would have used. https://www.census.gov/geographies/mapping-files/time-series/geo/tiger-line-file.2010.html#list-tab-790442341
- Select the file icon that says "FTP Archive by State"
- Select the state you are interested in. For us, this will be mostly Illinois, with a few tracts from Wisconsin and Indiana. Here is the link for Illinois: https://www2.census.gov/geo/pvs/tiger2010st/17_Illinois/
- Choose the file with the FIPS code you are interested in (such as 17031): https://www2.census.gov/geo/pvs/tiger2010st/17_Illinois/17031/
- Download the file called `tl_2010_17031_tract10.zip`. This file name stands for "TIGER/Line" (tl), the year (2010), the FIPS area (17031), and that we are looking at census tracts (tract) again for the year 2010 (10): https://www2.census.gov/geo/pvs/tiger2010st/17_Illinois/17031/
