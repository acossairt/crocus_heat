## Public Health Indicators of Susceptibility
This notebook provides data and a workflow to visualize a few population-level health statistics which may be related to *susceptibility* to heat-illness/mortality. Other factors may be important as well -- feel free to explore them! The data in this notebook primarily comes from the Chicago Health Atlas: https://chicagohealthatlas.org/, with data filtered by "Indicators."

**Why does this matter?**

In the path from weather to heat-related illness/mortality, *susceptibility* includes personal factors (especially health factors) which increase an individual's risk of an adverse health outcome when exposed to heat. For instance, it is well known that human thermoregulatory response is weaker in elderly individuals and very young children compared to younger adults. Additionally, individuals may have suppressed thermoregulatory capacity if they have diabetes, underlying cardiovascular diseases (such as hypertension), take certain medications, etc.

**Interpretation**

It can be tricky to differentiate between *susceptibility* and individual barriers to *adaptive capacity*. For instance, individuals with limited mobility or who are experiencing psychological distress have historically fared worse in extreme heat conditions compared to the general population. Is that because these individuals' health conditions cause them to have a less effective thermoregulatory response? Or is it because their conditions limit their capacity to adapt to the heat by, for example, seeking shelter in a cool environment? In most cases, the answer is probably a combination of both. Ultimately, these distinctions between "susceptibility," "adaptive capacity," and other factors are arbitrary. We should use these labels in-so-far as they help us generate a useful model, but we should not trick ourselves by sticking too closely to these categories of risk-factors.

**Data**

I have provided pre-cleaned data from the Chicago Health Atlas for the following list of indicators. To see how I cleaned this data, check out `crocus_heat/datasets/chi_health_atlas_cleaning.ipynb`.
- Adult asthma rates
- Adult diabetes rates
- Adult loneliness rates
- Asthma emergency department visits
- Behavioral health related hospitalizations
- Behavioral health treatment rates
- Diabetes related hospitalizations
- Hypertension rates
- Mood-disorder related hospitalizations
- Serious Psychological Distress (reported rates)
- Scizophrenic Disorder related hospitalizations

```python
import numpy as np
import pandas as pd
import geopandas as gpd
from matplotlib import pyplot as plt

health_data_path = "./datasets/chi_health_atlas_clean/"
boundaries_data_path = "./datasets/boundaries_geospatial_clean/"
commareas = gpd.read_file(boundaries_data_path + "chi_commareas_2018_clean.shp")
zips = gpd.read_file(boundaries_data_path + "chi_zip_2020_clean.shp")
```

### Adult diabetes -- example of statistics by community area
Source info available at the following links (click "Sources")
- https://chicagohealthatlas.org/indicators/HCSDIA?topic=adult-diabetes
- https://chicagohealthatlas.org/indicators/HCSDIAP?topic=adult-diabetes-rate

```python
diabetes = pd.read_csv(health_data_path + "adult_diabetes_commareas.csv")

# Some health datasets do not have data for every community area
print("Rows in health dataset:", len(diabetes))
print("Rows in boundary dataset:", len(commareas))

# Create a shape object for health info in each community area
diabetes_shp = pd.merge(left=commareas, right=diabetes, how="outer", left_on=["community", "area_num"], right_on=["community", "area_num"])
diabetes_shp.head()
```

```python
# Plot diabetes rate by community area
ax = commareas.boundary.plot(edgecolor="black", linewidth=0.5, figsize=[5,10])
ax = diabetes_shp.plot(ax=ax, column="diabetes_rate", legend=True, cmap="viridis", legend_kwds={"shrink":0.35, "format":'%.2e'})
ax.set_title("Adult Diabetes Rate (2020 - 2021)")

# Note the community areas with highest / lowest rate
diabetes_max = np.nanmax(diabetes_shp["diabetes_rate"])
diabetes_min = np.nanmin(diabetes_shp["diabetes_rate"])
commarea_max = diabetes_shp.loc[diabetes_shp["diabetes_rate"] == diabetes_max, "community"].item()
commarea_min = diabetes_shp.loc[diabetes_shp["diabetes_rate"] == diabetes_min, "community"].item()

textstr1 = "Highest rate:\n" + str(commarea_max) + \
           "\nLowest rate:\n" + str(commarea_min)
props = dict(boxstyle='round', facecolor='blue', alpha=0.1)
ax.text(0.05, 0.25, textstr1, transform=ax.transAxes, fontsize=10,
        verticalalignment='top', bbox=props)

plt.savefig("./plots/adult_diabetes_commareas.png", bbox_inches="tight", facecolor="white")
plt.show()
```

## Asthma Emergency Department Visits -- example of statistics by zip code
Source info available at the following links (click "Sources")
- https://chicagohealthatlas.org/indicators/HDED?topic=asthma-ed-visits
- https://chicagohealthatlas.org/indicators/HDEDR?topic=asthma-ed-visit-rate

```python
asthma_ed = pd.read_csv(health_data_path + "asthma_ed_visits_zip.csv")

# Some health datasets do not have data for every zip code
print("Rows in health dataset:", len(asthma_ed))
print("Rows in boundary dataset:", len(zips))

# Create a shape object for health info in each zip code
asthma_ed_shp = pd.merge(left=zips, right=asthma_ed, how="outer", left_on="zip", right_on="zip_code")
asthma_ed_shp.head()
```

```python
# Plot asthma emergency department visits by zip code
ax = zips.boundary.plot(edgecolor="black", linewidth=0.5, figsize=[5,10])
ax = asthma_ed_shp.plot(ax=ax, column="asthma_ed_rate", legend=True, cmap="viridis", legend_kwds={"shrink":0.35, "format":'%.2e'})
ax.set_title("Asthma Emergency Department Visits (2017)")

# Note the zip codes with highest / lowest rate
asthma_ed_max = np.nanmax(asthma_ed_shp["asthma_ed_rate"])
asthma_ed_min = np.nanmin(asthma_ed_shp["asthma_ed_rate"])
zip_max = asthma_ed_shp.loc[asthma_ed_shp["asthma_ed_rate"] == asthma_ed_max, "zip"].item()
zip_min = asthma_ed_shp.loc[asthma_ed_shp["asthma_ed_rate"] == asthma_ed_min, "zip"].item()

textstr1 = "Highest rate:\nzip code " + str(zip_max) + \
           "\nLowest rate:\nzip code " + str(zip_min)
props = dict(boxstyle='round', facecolor='blue', alpha=0.1)
ax.text(0.05, 0.25, textstr1, transform=ax.transAxes, fontsize=10,
        verticalalignment='top', bbox=props)

plt.savefig("./plots/adult_asthma_ed_zips.png", bbox_inches="tight", facecolor="white")
plt.show()
```
